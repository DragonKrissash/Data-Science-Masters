```python
def get_cols_zero_std(df : pd.DataFrame):
    cols_to_drop=[]
    num_cols=[i for i in df.columns if df[i].dtype !="O"]
    for i in df.columns[1:]:
        if df[i].std()==0:
            cols_to_drop.append(i)
    return cols_to_drop
```

```python
def get_reduntant_col(df:pd.DataFrame,missing_thresh=0.7):
    cols_missing_ratio=df.isnull().sum().div(df.shape[0])
    cols_to_drop=list(cols_missing_ratio[cols_missing_ratio>missing_thresh].index)
    return cols_to_drop
```

```python
from sklearn.cluster import KMeans
```

```python
from kneed import KneeLocator
from typing import Tuple
from dataclasses import dataclass
```

```python
@dataclass
class ClassDataInstances:
    X : np.array
    desc : str
    def _get_ideal_number_of_cluster(self):
        try:
            print(f'Getting the ideal number of Cluster {self.desc} set into ')
            print(f'Computing WCSS')
            wcss=[]
            for i in range(1,11):
                kmean=KMeans(n_clusters=i,init='k-means++',random_state=42)
                kmean.fit(self.X)
                wcss.append(kmean.inertia_)
                print(f'WCSS for n_clusters={i}:{kmean.inertia_}')
            print('WCSS computed successfully for all shortlisted number of clusters!')
            print('Finding the ideal number of clusters (by locating the dispersal point) via Elbow method')
            knee_finder=KneeLocator(
                range(1,11),wcss,curve='convex',direction='decreasing')
            print(f'Ideal number of clusters to be formed: {knee_finder.knee}')
            return knee_finder.knee
        except Exception as e:
            print(e)
            raise e

    def create_cluster(self)->Tuple:
        try:
            ideal_cluster=self._get_ideal_number_of_cluster()
            print(f'Dividing the \"{self.desc}\" instances into {ideal_cluster} cluster via KMeans Cluster' )
            kmeans=KMeans(n_clusters=ideal_cluster,init='k-means++',random_state=42)
            y_kmean=kmeans.fit_predict(self.X)
            print(f'{ideal_cluster} created successfully')
            return kmeans,np.c_[self.X,y_kmean]
        except Exception as e:
            print(e)
            raise e
             
```

```python
cluster_mast=ClassDataInstances(X=X_trans,desc="Wafers features")
cluster,X_clus=cluster_mast.create_cluster()
```
