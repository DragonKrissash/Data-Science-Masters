```python
def corr_heatmap(data,idx_s,idx_e):
    y=data['labels']
    temp=data.iloc[:,idx_s:idx_e]
    if 'id' in temp.columns:
        del temp['id']
    temp['labels']=y
    sns.heatmap(temp.corr(),annot=True,fmt='.2f')
    plt.show()
```

```python
from sklearn.feature_selection import mutual_info_classif
```

```python
X=df.drop('labels',axis=1)
```

```python
y=df['labels']
```

```python
discrete_features=X.dtypes==int
```

```python
mi_score=mutual_info_classif(X,y,discrete_features=discrete_features)
```

```python
mi_score=pd.Series(mi_score,name='MI score',index=X.columns)
mi_score=mi_score.sort_values(ascending=False)
mi_score
```

```python
def plot_mi_score(scores):
    scores=scores.sort_values(ascending=True)
    width=np.arange(len(scores))
    ticks=list(scores.index)
    plt.barh(width,scores)
    plt.yticks(width,ticks)
    plt.title('MI Scores')

plt.figure(dpi=100,figsize=(12,12))
plot_mi_score(mi_score)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
```

```python
def train_logistic(data,top_n):
    top_n_fea=mi_score.sort_values(ascending=False).head(top_n).index.tolist()
    X=data[top_n_fea]
    y=data['labels']
    Xtr,Xte,ytr,yte=tts(X,y,random_state=42,train_size=0.2,shuffle=True)
    lr=LogisticRegression()
    lr.fit(Xtr,ytr)
    y_pred=lr.predict(Xte)
    precision=precision_score(y_pred,yte)
    recall=recall_score(y_pred,yte)
    f1=f1_score(y_pred,yte)
    accuracy=accuracy_score(y_pred,yte)
    return precision,recall,f1,accuracy
```

```python
arr=[]
for i in range(20,51):
    precision,recall,f1,accuracy=train_logistic(df,i)
    print(f'Performance for Logistic model with Top {i} features is : precision: {precision} , recall: {recall} , f1: {f1} , accuracy: {accuracy}')
    arr.append([i,precision,recall,f1,accuracy])
```

```python
df_log=pd.DataFrame(arr,columns=['No of columns','precision','recall','f1_score','accuracy'])
df_log
```

```python
df_log['No of columns'][df_log['recall'].idxmax()]
```

```python
sns.lineplot(x='No of columns',y='precision', data=df_log,label='Precison Score')
sns.lineplot(x='No of columns',y='recall', data=df_log,label='Recall Score')
sns.lineplot(x='No of columns',y='f1_score', data=df_log,label='F1 Score')
sns.lineplot(x='No of columns',y='accuracy', data=df_log,label='Accuracy Score')
```

```python
def train_rfc(data,top_n):
    top_n_fea=mi_score.sort_values(ascending=False).head(top_n).index.tolist()
    X=data[top_n_fea]
    y=data['labels']
    Xtr,Xte,ytr,yte=tts(X,y,random_state=42,train_size=0.2,shuffle=True)
    rfc=RandomForestClassifier(n_estimators=100,criterion='entropy',max_depth=32,max_features=1.0)
    rfc.fit(Xtr,ytr)
    y_pred=rfc.predict(Xte)
    precision=precision_score(y_pred,yte)
    recall=recall_score(y_pred,yte)
    f1=f1_score(y_pred,yte)
    accuracy=accuracy_score(y_pred,yte)
    return precision,recall,f1,accuracy
```

```python
arr_rfc=[]
for i in range(20,51):
    precision,recall,f1,accuracy=train_rfc(df,i)
    print(f'Performance for RFC model with Top {i} features is : precision: {precision} , recall: {recall} , f1: {f1} , accuracy: {accuracy}')
    arr_rfc.append([i,precision,recall,f1,accuracy])
```

```python
df_rfc=pd.DataFrame(data=arr_rfc,columns=['No of columns','precision','recall','f1_score','accuracy'])
```

```python
sns.lineplot(x='No of columns',y='precision', data=df_rfc,label='Precison Score')
sns.lineplot(x='No of columns',y='recall', data=df_rfc,label='Recall Score')
sns.lineplot(x='No of columns',y='f1_score', data=df_rfc,label='F1 Score')
sns.lineplot(x='No of columns',y='accuracy', data=df_rfc,label='Accuracy Score')
```
